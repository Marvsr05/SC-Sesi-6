```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda t, s: np.exp(-t)
h = 0.1
t_min = 0
t_max = 1

t = np.arange(t_min, t_max + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t)-1):
    s[i+1] = s[i] + h * f(t[i], s[i])

plt.figure(figsize=(12, 8))
plt.plot(t, s, 'go--', label='approx (euler h = 0.1)')
plt.plot(t, -np.exp(-t), label = 'exact solution')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f1 = lambda t, s: t**2 - 4*s
h = 0.1
t_min = 0
t_max = 1
t = np.arange(t_min, t_max + h, h)
s0 = 1

s = np.zeros(len(t))
s[0] = s0
for i in range(0, len(t)-1):
    s[i+1] = s[i] + h * f1(t[i], s[i])

plt.figure(figsize=(10, 6))
plt.plot(t, s, 'bo--')
plt.title("Solusi: $y' + 4y = x^2$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f2 = lambda t, s: np.sin(s)
h = 0.1
t_min = 0
t_max = 1
t = np.arange(t_min, t_max + h, h)
s0 = 1

s = np.zeros(len(t))
s[0] = s0
for i in range(0, len(t)-1):
    s[i+1] = s[i] + h * f2(t[i], s[i])

plt.figure(figsize=(10, 6))
plt.plot(t, s, 'ro--')
plt.title("Solusi: $y' = \sin(y)$")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
